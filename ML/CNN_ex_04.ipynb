```python
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# set random seed
tf.random.set_seed(42)

# process the image and data normalization
train_data_gen = ImageDataGenerator(rescale=1./255)
test_data_gen = ImageDataGenerator(rescale=1./255)

# train/test dir
train_dir = "/content/pizza_steak/train"
test_dir = "/content/pizza_steak/test"

# import train data/ test data from dir and turn into batches
train_data = train_data_gen.flow_from_directory(
                                              directory=train_dir, 
                                              batch_size=32, 
                                              target_size=(224,224), 
                                              seed=42,
                                              class_mode="binary"
                                              )

test_data = train_data_gen.flow_from_directory(
                                              directory=test_dir, 
                                              batch_size=32, 
                                              target_size=(224,224), 
                                              seed=42,
                                              class_mode="binary"
                                              )
```

```python
# creating a CNN model
cnn_model1 = tf.keras.Sequential([
        tf.keras.layers.Conv2D(
                              name='Layer_01',
                              filters=10,
                               kernel_size = 3,
                               activation='relu',
                               input_shape=(224, 224, 3)),
        tf.keras.layers.Conv2D(10,3, activation='relu', name='Layer_02'),
        tf.keras.layers.MaxPool2D(pool_size=2, padding='valid', name='Layer_03'),
        tf.keras.layers.Conv2D(10,3, activation='relu', name='Layer_04'),
        tf.keras.layers.Conv2D(10,3,activation='relu', name='Layer_05'),
        tf.keras.layers.MaxPool2D(2, name='Layer_06'),
        tf.keras.layers.Flatten(name='Layer_07'),
        tf.keras.layers.Dense(1, activation='sigmoid',name='final_layer')
])

# compile the model

cnn_model1.compile(
    loss=tf.keras.losses.binary_crossentropy,
    optimizer = tf.keras.optimizers.Adam(),
    metrics=["accuracy"]
)

# fit the model
hist_1 = cnn_model1.fit(
                    train_data,
                    epochs=5,
                    steps_per_epoch=len(train_data),
                    validation_data=test_data,
                    validation_steps=len(test_data)
                    )
```
