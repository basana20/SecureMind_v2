```python
import tensorflow as tf
import numpy as np
```

```python
xs = np.array([-7.0, -5.0, -3.0, 1.0, 3.0, 5.0])
ys = np.array([3.0, 5.0, 7.0, 11.0, 13.0, 15.0])
```

```python
tf.random.set_seed(42)
x_tf = tf.cast(tf.constant(xs), tf.float32)
y_tf = tf.cast(tf.constant(ys), tf.float32)
```

```python
x_tf, y_tf
```

```python
# creating a sample model
model = tf.keras.Sequential([
tf.keras.layers.Dense(1)
])

# compile a model
model.compile(
    loss = tf.keras.losses.mae,
    optimizer= tf.keras.optimizers.SGD(),
    metrics=["mae"]
)

# fitting a model with changing the no.of epochs to get the better output
model.fit(x_tf,y_tf, epochs=10)
```

```python
x_tf, y_tf
```

```python
model.predict([7])
```

```python
# creating s new model 
model2 = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(1)
])
# create new complier
model2.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.01),
    loss = tf.keras.losses.mae,
    metrics=["mae"]
)

# fitting a model
model2.fit(x_tf, y_tf, epochs=100)
```

```python
model2.predict([20])
```

```python
model_xs2 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(100,activation='relu', input_shape = [1], name='Input_Layer'),
    tf.keras.layers.Dense(1)
    # tf.keras.layers.Dense(4,activation='relu', name='second_dense_layer'),
    # tf.keras.layers.Dense(4,activation='relu', name='third_dense_layer')
])
# compile a model_xs2
model_xs2.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.001),
    loss = tf.keras.losses.mae,
    metrics=["mae"]
)
```

```python
model_xs2.summary()
```

```python
# splitting the data for tranning and testnig on xs2
xs2_train = xs2[:60]
xs2_test = xs2[60:]
xs2_train, xs2_test
```

```python
# splitting the data for tranning and testnig on ys2
ys2_train = ys2[:60]
ys2_test = ys2[60:]
ys2_train, ys2_test
```

```python
model_xs2.fit(xs2_train, ys2_train, epochs=500)
```

```python
ys2_predict = model_xs2.predict(xs2_test)
ys2_predict
```

```python
import matplotlib.pyplot as plt

def plot_prediction(x_train, y_train, x_test, y_test, x_pred, y_pred):
  # to visulize the data between train,test and predict
  plt.figure(figsize=(15,10))
  # plot tranning data
  plt.scatter(x_train, y_train, label='Trained Data', c='b') 
  # plot test data
  plt.scatter(x_test, y_test, label='Tested Data', c='g')
  # plot predicted data
  plt.scatter(x_pred, y_pred, label='predicted Data', c='r') 
  # show the data
  plt.legend();
```

```python
plot_prediction(
    x_train=xs2_train, 
    y_train=ys2_train,
     x_test=xs2_test, 
     y_test=ys2_test, 
     x_pred=xs2_test, 
     y_pred=ys2_predict
     )
```

```python
# finding mean absolute Error
mae = tf.metrics.mean_absolute_error(y_true=ys2_test, y_pred=tf.squeeze(ys2_predict))
mae
```

```python
# finding mean absolute square error
mse = tf.metrics.mean_squared_error(y_true=ys2_test, y_pred=tf.squeeze(ys2_predict))
mse
```
