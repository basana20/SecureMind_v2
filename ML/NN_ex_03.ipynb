```python
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# read CSV file via pandas
insurance_data = pd.read_csv("https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv")
insurance_data
```

```python
insurance_data.head()
```

```python
# get dummies to pandas data sets
insurance_data_dummies = pd.get_dummies(insurance_data)
```

```python
insurance_data_dummies.head()
```

```python
X = insurance_data_dummies.drop('charges', axis=1)
y = insurance_data_dummies['charges']
```

```python
X.head(), y.head()
```

```python
# for spliting the data using sklearn.model_selection.train_test_split
from sklearn.model_selection import train_test_split
```

```python
 # get the data using the below method
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
len(X_train), len(y_test)
```

```python
X_train.head(), X_test.head()
```

```python
tf.random.set_seed(42)
# create a NN model 
insurance_model = tf.keras.Sequential([
      tf.keras.layers.Dense(100, name='first_layer'),
      tf.keras.layers.Dense(10, name='second_layer'),
      tf.keras.layers.Dense(1)
])

# compile the insurance model
insurance_model.compile(
    loss= tf.keras.losses.mae,
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
    metrics = ['mae']
)
# fitting a insurance_model
history = insurance_model.fit(X_train, y_train, epochs=200, verbose=0)
```

```python
insurance_model.evaluate(X_test, y_test)
```

```python
pd.DataFrame(history.history).plot()
plt.ylabel('loss')
plt.xlabel('epochs')
```
